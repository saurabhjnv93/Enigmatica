THE BELOW CODE IS TO MAKE TOPIC WISE ANALYSIS AS NOW WE DONT HAVE ENOGH DATA. IF WE HAVE ENOUGH DATA THEN WE WILL MAKE CLASSIFY THE TAGS OF THE QUESTION AND ASSIGN THAT PART TO THE PARTICULAR QUESTION. SO THAT ACCESING ALSO EASY AND GET RELEVENT INFORMATION 

```python
import pandas as pd
from bertopic import BERTopic




topic_model = BERTopic(language="english", verbose=True)

# Fit BERTopic on paragraphs
topics, probs = topic_model.fit_transform(df["paragraph"])

# Add topics to the DataFrame
df["topic"] = topics

# Display topic information
print("Topic Overview:")
print(topic_model.get_topic_info())

# Manual or Automated Labeling
# Extract the top terms for each topic
topic_labels = {}
for topic_id in df["topic"].unique():
    if topic_id == -1:
        topic_labels[topic_id] = "Miscellaneous"
    else:
        topic_terms = topic_model.get_topic(topic_id)
        top_words = ", ".join([word[0] for word in topic_terms[:5]])
        # Optional: Assign a descriptive label based on top terms
        topic_labels[topic_id] = f"Topic: {top_words}"

# Map topics to labels
df["topic_label"] = df["topic"].map(topic_labels)

# Save results to a CSV file
df.to_csv("labeled_paragraphs.csv", index=False)

# Print labeled DataFrame
print("\nLabeled DataFrame:")
print(df)

# Visualize topics
topic_model.visualize_topics()
```

```python
import pandas as pd

# Load the datasets
df1 = pd.read_csv("../ML/testing1.csv")
df2 = pd.read_csv("../ML/testing2.csv")

# Combine the DataFrames
combined_df = pd.concat([df1, df2], ignore_index=True)

# Display the first few rows of the combined DataFrame
combined_df
```

```python
# Import the Sentence Transformer library
from sentence_transformers import SentenceTransformer, util

def sim(s1, s2):
    model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

    # Encode the sentences into embeddings
    sent_embedding1 = model.encode(s1, convert_to_tensor=True)
    sent_embedding2 = model.encode(s2, convert_to_tensor=True)

    # Compute the cosine similarity and extract the scalar value
    similarity = util.pytorch_cos_sim(sent_embedding1, sent_embedding2).item()
    return similarity
```

```python
def some_function(x, y):
    # Example: Concatenate strings with a separator
    return sim(x,y) 
```
